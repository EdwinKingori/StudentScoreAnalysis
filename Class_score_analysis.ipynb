### Using different distribution plots to visualize data

#### The distribution of the math_score

```python
sns.set_theme( style = 'darkgrid', font_scale = 1.2)
sns.displot(student_df['math_score'], kde = True, bins =  125)
```

```python
sns.histplot(student_df['math_score'], kde = False)
```

#### The distribution of maths_core across 'gender'.

```python
# example 1: using a histplot
sns.histplot(student_df, x = 'math_score', hue = "gender", kde = True, palette =['lightpink', 'darkblue'])
```

#### Distribution of maths_core across 'ethnicity'

```python
# solution 1: using catplot
#using the categorical catplot of the boxplot kind to visualize the distribution maths_core across 'ethnicity'
sns.catplot(student_df, x = 'math_score', y = 'ethnicity', kind = 'box' , hue = 'gender', palette = ['pink', 'blue'])
```

```python
#solution 2: Using the KDE plot
sns.kdeplot(student_df, x = 'math_score', hue = 'ethnicity', common_norm = False)
```

### Relationship between the maths_core and reading_score variables

```python
student_df.head()
```

```python
#Using the scatterplot, which is the best for visualizing the relationship between two numeric variables
sns.scatterplot(student_df, x = 'math_score', y = 'reading_score', hue = 'gender', palette = ['orange', 'blue'])
```

```python
# using a boxplot to show the comparison between math_score and test_preparation_course
sns.boxplot(student_df, x = 'math_score', y = 'test_preparation_course', hue = 'gender', palette = ['pink', 'blue'] )
```

Note: The barline that goes accross he middle of the boxes is the median, and the box extends one_standard deviation(that's what the box represents) from the median. The lines above the boxes are whiskers. And they extend to all of the other data asides from what is in the standaed deviation. The star-like shapes are considered to be outliers.

```python
sns.pairplot(student_df, hue = 'gender', palette = ['brown', 'blue'])
```
